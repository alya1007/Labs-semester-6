```python
import matplotlib.pyplot as plt
import numpy as np

Ts = 0.01
t = np.arange(0, 5, Ts)
x1 = np.random.normal(size=len(t))

plt.plot(t, x1)
plt.grid(True)
plt.title('Gaussian white noise')
plt.xlabel('time(s)')
plt.ylabel('function y(t)')
plt.show()
```

# 1.5

```python
from scipy.signal import lfilter

Ts=0.01
om0 = 2 * np.pi
dz = 0.005
A = 1

oms = om0 * Ts

a = [1 + 2 * dz * oms + oms ** 2, -2 * (1 + dz * oms), 1]
b = [A * 2 * oms ** 2]

t = np.arange(0, 50, Ts)
x1 = np.random.rand(len(t))
y1 = lfilter(b, a, x1)

plt.plot(t, y1)
plt.grid(True)
plt.title('Second order low-pass filter')
plt.xlabel('time(s)')
plt.ylabel('function y(t)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter


Ts = 0.001
om0 = 2 * np.pi
dz = 0.005
A = 1

oms = om0 * Ts

a = [1 + 2 * dz * oms + oms ** 2, -2 * (1 + dz * oms), 1]
b = [A * 2 * oms ** 2]

t2 = np.arange(0, 50, Ts)
x2 = np.random.rand(len(t2))
y2 = lfilter(b, a, x2)

plt.plot(t2, y2)
plt.grid(True)
plt.title('Second order low-pass filter')
plt.xlabel('time(s)')
plt.ylabel('function y(t)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

R = 50
m = np.arange(0, R)
s = 2 * m * (0.9 ** m)

d = np.random.rand(len(m)) - 0.5

s_noisy = s + d

plt.plot(m, s, label='Clear signal')
plt.plot(m, s_noisy, label='Signal with noise')
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Clear signal and signal with noise')
plt.legend()
plt.show()
```

# 2.5

```python
R = 50
m = np.arange(0, R)
s = 2 * m * (0.9 ** m)

d = np.random.rand(len(m)) - 0.5

x = s + d

M = 3
b = np.ones(M) / M

y = lfilter(b, 1, x)

plt.plot(m, s, label='Clear signal', color='blue')
plt.plot(m, x, label='Signal with noise', color='green')
plt.plot(m, y, label='Filtered signal', color='red')
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Clear signal, signal with noise and filtered signal')
plt.legend()
plt.show()
```

# 2.6

```python
R = 50
m = np.arange(0, R)
s = 2 * m * (0.9 ** m)

d = np.random.rand(len(m)) - 0.5

x = s + d

M_5 = 5
b_5 = np.ones(M_5) / M_5
y_5 = lfilter(b_5, 1, x)

M_10 = 10
b_10 = np.ones(M_10) / M_10
y_10 = lfilter(b_10, 1, x)

plt.plot(m, s, label='Clear signal', color='blue')
plt.plot(m, x, label='Signal with noise', color='green')
plt.plot(m, y_5, label='Filtered signal (MAF, M=5)', color='red')
plt.plot(m, y_10, label='Filtered signal (MAF, M=10)', color='orange')
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Clear signal, signal with noise and filtered signals')
plt.legend()
plt.show()
```

# 2.7

```python
from scipy.signal import lfilter, sawtooth

R = 1
m = np.arange(0, R, 0.001)
s = 2 * sawtooth(3 * np.pi * m + np.pi/6)

d = np.random.rand(len(m)) - 0.5

x = s + d

M = 20
b = np.ones(M) / M
y = lfilter(b, 1, x)

plt.plot(m, s, label='Clear signal', color='blue')
plt.plot(m, x, label='Signal with noise', color='green')
plt.plot(m, y, label='Filtered signal (MAF, M=20)', color='red')
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Clear signal, signal with noise and filtered signal (MAF, M=20)')
plt.legend()
plt.show()
```
